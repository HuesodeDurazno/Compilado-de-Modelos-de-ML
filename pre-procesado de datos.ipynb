```python
dataset = pd.read_csv('/content/machinelearning-az/datasets/Part 1 - Data Preprocessing/Section 2 -------------------- Part 1 - Data Preprocessing --------------------/Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 3].values
display(dataset)
```

# Tratamiento de los NAs

importa funcion simpleimputer de sklearn de impute (impute ->toma una base de datos y decide que hacer con los datos faltantes)

```python
from sklearn.impute import SimpleImputer
```

Buscamos campiar los valores faltantes por los valores medios.

primer valor ->valores a conciderar en este caso "nan" representados por np.nan

segundo par-> estrategia a implementar (media)

tercero->controla la velocidad del imputador ??' xd default 0

fit es llamada para aplicar la funcion simpleimputer , marcamos que solo queremos todas las coulmnas pero columnas de la 1->3

transform->sobreescribe los datos en nuestros datos X

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean", verbose=0)
imputer = imputer.fit(X[:,1:3]) 
X[:, 1:3] = imputer.transform(X[:,1:3])
#codigo agregado para ver que esta pasando , no se asusten raza
df=pd.DataFrame(X)
display(df)
```

# Codificar datos categoricos

siguiente paso convertir datos categoricos a su representacion numerica 

importamos librerias

declaramos un encoder que mandara a llamar la funcion .fit_transform sobre la columna 0 para transformar todos estos datos categoricos a su representacion numerica, surge un problema , datos ordinales que representan una jerarquia lo cual no es correcto.

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
#imprimimos
df=pd.DataFrame(X)
display(df)
```

Transformamos estos datos ordinales a variables dummy ,(forma de traducir una categoria que no tiene un orden), se generaran 3 columnas , una para cada categoria ,donde se encontrara un 1 en caso de representar dicha categoria y 0 en caso de no hacerlo

ColumTransformer->realiza varias transformaciones a 1 o varias columnas

  par 1->Lista de tuplas (nombre, transformador, columnas) que especifican los objetos transformadores que se aplicarán a subconjuntos de datos.Onehotencoder resive por default auto determina automaticamente las categorias 

  par2-> remainder por default eliminara todas las columnas no transformadas, si indicamos passthrough se pasaran aquellas no transformadas

  "buscar transformaciones posibles"

```python
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [0])],   
    remainder='passthrough'                        
)
```

convertimos a array la aplicacion del transformador , e indicamos que es de tipo float, lo guardamos en X

aplicamos solo un lavelencoder a y pues son valores booleanos , previamente explicado

```python
X = np.array(ct.fit_transform(X), dtype=np.float)
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
df=pd.DataFrame(X)
display(df)
df=pd.DataFrame(y)
display(df)
```

# Dividir el data set en conjunto de entrenamiento y conjunto de testing


importamos libreria que nos ayudara a dividir nuestro conjunto de entranmiento y testing

train_test_split devuelve una lista que contiene los conjunto de entrenamiento entrenamiento y pruebas de las matrices que indiquemos en paremostros

pasamos matriz x y y indicamos la longitud del test que sera 20%

numero cualquiera que reordena los datos al azar, el random_state es una semilla que nos permite reproducir el mismo ordenamiento al azar

queremos el reordenamiento de datos para el algoritmo no aprenda un sesgo(patron)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
df=pd.DataFrame(X_train)
display(df)
df=pd.DataFrame(X_test)
display(df)
df=pd.DataFrame(y_train)
display(df)
df=pd.DataFrame(y_test)
display(df)
```
