# Natural Language Processing

# Cómo importar las librerías


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importar el data set


leemos el documento indicando que el separador sera un tabulador y tambien ignoramos las comillas con quoting=3

```python
dataset = pd.read_csv("/content/machinelearning-az/datasets/Part 7 - Natural Language Processing/Section 36 - Natural Language Processing/Restaurant_Reviews.tsv", delimiter = "\t", quoting = 3)
```

```python
display(dataset)
```

# Limpieza de texto

```python
import re#libreria que nos permite ocupar expresiones regulares
import nltk#libreria para stopwords
nltk.download('stopwords')#descargamos stop words
from nltk.corpus import stopwords#importamos las stopwords
from nltk.stem.porter import PorterStemmer
corpus = []#guardamos las oraciones limpias en este arreglo
for i in range(0, 1000):
    review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])#subtituimos todos lo que no este entre a y z y A y Z por " " en todas las horaciones 
    review = review.lower()#convertimos a minusculas
    review = review.split()#separamos las palabras por espacio
    ps = PorterStemmer()#nos ayuda a lematizar las palabras
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]#ciclo que va guardando las palabras diferentes a stop words 
    review = ' '.join(review)#unimos la cadena
    corpus.append(review)#agregamos al corpus
```

# Crear el Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer#nos permite hacer la bolsa de palabras
cv = CountVectorizer(max_features = 1500)#hacemos volsa de palabras con las 1500 palabras mas relevantes
X = cv.fit_transform(corpus).toarray()#convertimos la bolsa de palabras a array
y = dataset.iloc[:, 1].values#sacamos nuestro y
```

# Dividir el data set en conjunto de entrenamiento y conjunto de testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

# Ajustar el clasificador en el Conjunto de Entrenamiento

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

# Predicción de los resultados con el Conjunto de Testing

```python
y_pred  = classifier.predict(X_test)
```

# Elaborar una matriz de confusión

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(cm)
(55+91)/200
```
