# Parte 1 - Construir el modelo de CNN# Parte 1 - Pre procesado de datos

# Importar las liobrerías y paquetes

```python
from keras.models import Sequential#inicializar los pesos de la red
from keras.layers import Conv2D#generar convolucion 2d
from keras.layers import MaxPooling2D#generar max pooling 2d
from keras.layers import Flatten#generar flatten
from keras.layers import Dense#crea las capas densamente conectadas
```

# Inicializar la CNN

```python
classifier = Sequential()#iniciamos la red neuronal
```

# Paso 1 - Convolución

```python
classifier.add(Conv2D(filters = 32,kernel_size = (3, 3), 
                      input_shape = (64, 64, 3), activation = "relu"))
#cantidad de filtros,tamaño del kernel , input shape indica tamaño de la imagen y canales de color, y activacion es la funcion de activacion en este caso el rectificador lineal unitario
```

# Paso 2 - Max Pooling

```python
classifier.add(MaxPooling2D(pool_size = (2,2)))
#agregamos la ventana max pooling de 2x2
```

# Una segunda capa de convolución y max pooling

```python
classifier.add(Conv2D(filters = 32,kernel_size = (3, 3), activation = "relu"))
```

```python
classifier.add(MaxPooling2D(pool_size = (2,2)))
```

# Paso 3 - Flattening

```python
classifier.add(Flatten())
#agregamos capa de flatten
```

# Paso 4 - Full Connection

```python
classifier.add(Dense(units = 128, activation = "relu"))
classifier.add(Dense(units = 1, activation = "sigmoid"))
#agregamos dos capas densamente conectadas una con 128 neuronas con una funcion de activacion de rectificador lineal unitario y otra de salida con funcion sigmoide
```

# Compilar la CNN

```python
classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
#compilamos la red , con optimizador adam, funcion de perdida y metricas de precision que en este caso seria precision xd
```

# Parte 2 - Ajustar la CNN a las imágenes para entrenar 

```python
from keras.preprocessing.image import ImageDataGenerator
#importamos la libreria para el preprocesado de imagenes
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
#transofrmaciones que se realizan a las imagenes ,rescalado de imagenes(normalizando 0-1),redimencion horizontal de la imagen, zom y giro de la imagen 
test_datagen = ImageDataGenerator(rescale=1./255)
#rescalamos la imagen
```

```python
#guardamos nuetro training y testing fataset, indicamo ruta, tamaño de la imagen, en grupos de cuantas imagenes tomara la red,tipo de clasificacion
training_dataset = train_datagen.flow_from_directory('/content/machinelearning-az/datasets/Part 8 - Deep Learning/Section 40 - Convolutional Neural Networks (CNN)/dataset/training_set',
                                                    target_size=(64, 64),
                                                    batch_size=32,
                                                    class_mode='binary')
```

```python
testing_dataset = test_datagen.flow_from_directory('/content/machinelearning-az/datasets/Part 8 - Deep Learning/Section 40 - Convolutional Neural Networks (CNN)/dataset/test_set',
                                                target_size=(64, 64),
                                                batch_size=32,
                                                class_mode='binary')
```

# Elaborar una matriz de confusión

```python
classifier.fit_generator(training_dataset,
                        steps_per_epoch=8000,
                        epochs=25,
                        validation_data=testing_dataset,
                        validation_steps=2000)
```
