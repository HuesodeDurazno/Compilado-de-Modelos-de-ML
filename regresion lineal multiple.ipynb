# Construir el modelo óptimo de RLM utilizando la Eliminación hacia atrás

agreamos una columna de 1 que nos ayudara a encontrar el p valor de las variables independientes


creamos el arrelo arr con una fila y 50 columnas del tipo entero , y a este le agreamos los valores del array x , esto se lo asignamos a x,axis=1 si se agrega en columnas y 0 si se agrega en filas 

statsmodels.api nos hayudara a utilizar eliminacion hacia atras

```python
import statsmodels.api as sm
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)
SL = 0.05
```

#Se ha añadido el modificador .tolist() al X_opt para adaptarse a Python 3.7

sm.OLS hace una regresion multiple pero nos retorna con sumario los valores p para procedeer a la eliminacion, endog son las variables a predecir y exog son las variables tomadas en cuenta en el modelo, es necesario pasar esto como una lista y no un array.

```python
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
regression_OLS = sm.OLS(endog = y, exog = X_opt.tolist()).fit()
regression_OLS.summary()

X_opt = X[:, [0, 1, 3, 4, 5]]
regression_OLS = sm.OLS(endog = y, exog = X_opt.tolist()).fit()
regression_OLS.summary()

X_opt = X[:, [0, 3, 4, 5]]
regression_OLS = sm.OLS(endog = y, exog = X_opt.tolist()).fit()
regression_OLS.summary()

X_opt = X[:, [0, 3, 5]]
regression_OLS = sm.OLS(endog = y, exog = X_opt.tolist()).fit()
regression_OLS.summary()

X_opt = X[:, [0, 3]]
regression_OLS = sm.OLS(endog = y, exog = X_opt.tolist()).fit()
regression_OLS.summary()
```

# Funciones extra para automatizar la regresion hacia atras y tomar en cuenta el valor de R cuadrado ajustado

funcion que recibe nuestro arreglo donde iremos eliminando filas y nuestro signification level

x es un array , tomamos la longitud de su primer fila que es igual al numero de columnas 

cremaos el modelo de regresion , tomamos el maximo de los valores p y si esta valor es mayor al valor a nuestro sl, lo buscamos dentro de nuestro array y eliminamos esa columna



```python
import statsmodels.api as sm
def backwardElimination(x, sl):    
    numVars = len(x[0])    
    for i in range(0, numVars):        
        regressor_OLS = sm.OLS(y, x.tolist()).fit()        
        maxVar = max(regressor_OLS.pvalues).astype(float)        
        if maxVar > sl:            
            for j in range(0, numVars - i):                
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):                    
                    x = np.delete(x, j, 1)    
    regressor_OLS.summary()    
    return x 
 
SL = 0.05
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_Modeled = backwardElimination(X_opt, SL)
```
